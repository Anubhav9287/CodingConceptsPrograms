## Minimum Island Size Problem
## Sample:



```python
grid = [
    ['W','L','W','W','W'],
    ['W','L','W','W','W'],
    ['W','W','W','L','W'],
    ['W','W','L','L','W'],
    ['L','W','W','L','L'],
    ['L','L','W','W','W']
]
tmp_count = 0
```

```python
def countLand(grid):
    visited = []
    global tmp_count
    result = float('inf')
    for rindex,r in enumerate(grid):
        for cindex,c in enumerate(r):
            if explore(grid,rindex,cindex,visited):
                result = min(result,tmp_count)
                tmp_count = 0
    return result
```

```python
def explore(grid, rindex, cindex, visited):
    global tmp_count
    pos = str(rindex)+','+str(cindex)
    #'0,0'
    r_inbound = 0 <= rindex and rindex < len(grid)
    c_inbound = 0 <= cindex and cindex < len(grid[0])
    if not r_inbound or not c_inbound:
        return False
    if pos in visited:
        return False
    visited.append(pos)
    if grid[rindex][cindex] == 'W':
        return False
    if grid[rindex][cindex] == 'L':
        tmp_count = tmp_count + 1
    explore(grid,rindex-1,cindex,visited)
    explore(grid,rindex+1,cindex,visited)
    explore(grid,rindex,cindex-1,visited)
    explore(grid,rindex,cindex+1,visited)
    return tmp_count
```

```python
countLand(grid)
```
