## Find the Shortest path


```python
edges = [
    ['w','x'],
    ['x','y'],
    ['z','y'],
    ['z','v'],
    ['w','v']
]
```

```python
#Firstly we will convert the given edges in adj form
def edgesToAdj(edges):
    adj = {}
    for edge in edges:
        if edge[0] not in adj:
            adj[edge[0]] = []
        if edge[1] not in adj:
            adj[edge[1]] = []
        adj[edge[0]].append(edge[1])
        adj[edge[1]].append(edge[0])
    return adj
```

```python
#Sample when we convert edges to Adj
edgesToAdj(edges)
```

```python
#To calculate the shortest path I will use breath first search
visited = []
def shortestPath(graph,src_node,dest_node):
    dist = float('inf')
    queue = []
    queue.append([src_node,0])
    while(queue):
        current = queue.pop(0)
        if current[0] == dest_node:
             dist = min(current[1],dist)
        if current[0] in visited:
            continue
        visited.append(current[0])
        for i in graph[current[0]]:
            queue.append([i,current[1]+1])
    return dist
```

```python
#here I will call the function to calculate path
src = 'x'
dest = 'y'
graph = edgesToAdj(edges)
shortestPath(graph,src,dest)
```
