## Largest Components


```python
cc = {
    '0':['8','1','5'],
    '1':['0'],
    '5':['0','8'],
    '8':['0','5'],
    '2':['3','4'],
    '3':['2','4'],
    '4':['3','2']
}
```

```python
visited = []
temp_comp = 0
```

```python
def Connectedcomponents(graph):
    global temp_comp
    comp_count = 0
    for eachNode in graph:
        comp_count = max(comp_count,explore(graph,eachNode,visited))
        temp_comp = 0
    return comp_count
```

```python
def explore(graph, eachNode,visited):
    global temp_comp
    if eachNode in visited:
        return 0
    visited.append(eachNode)
    temp_comp += 1
    for n in graph[eachNode]:
        explore(graph,n,visited)
    return temp_comp
```

```python
Connectedcomponents(cc)
```
